# Unit price" refers to the price of a single unit of a product or service, typically expressed in a specific currency (e.g., dollars, euros) per unit (e.g., per item, per kilogram, per liter). It is the price that consumers pay for each individual item or unit of a product.

# "Lag price" is not a standard term in economics or business, but I assume we are referring to the previous price point or historical price data. In this context, the "lag price" could be the price of the same product or a similar product at a previous point in time.

# The relationship between unit price and lag price can provide insights into pricing dynamics, consumer behavior, and market trends. Here's what the relationship might depict:

# Price Trends: Comparing unit prices to lag prices over time can help identify trends in pricing. For example, if unit prices consistently increase following higher lag prices, it might indicate an upward price trend or inflationary pressure. Conversely, if unit prices decrease after higher lag prices, it could suggest a trend of price discounts or deflationary pressures.

# Price Elasticity: Analyzing changes in unit price relative to changes in lag price can provide information about price elasticity of demand. If small changes in lag price lead to proportionally larger changes in unit price, the product may have elastic demand, meaning consumers are sensitive to price changes.

# Consumer Behavior: The relationship between unit price and lag price can reflect how consumers respond to changes in pricing. If consumers adjust their purchasing behavior significantly in response to changes in unit price relative to lag price, it can indicate how price-sensitive they are and how willing they are to adapt their consumption based on price fluctuations.

# Market Competition: Observing how unit prices respond to changes in lag prices can provide insights into competitive dynamics. For example, if a business consistently adjusts its unit prices in response to changes in lag prices (potentially set by competitors), it may indicate a competitive pricing strategy aimed at maintaining market share.

# Product Differentiation: The relationship between unit price and lag price can reveal whether businesses are using price changes as a form of product differentiation. If a product's unit price consistently increases following higher lag prices, it might signal an attempt to position the product as premium or of higher quality.

# Demand Forecasting: Analyzing how unit prices change in relation to lag prices can assist in demand forecasting. If unit prices consistently rise following higher lag prices and consumer behavior remains stable, it might indicate that demand for the product is relatively inelastic.

# Psychological Pricing: The relationship between unit price and lag price can provide insights into psychological pricing effects. Businesses may use certain pricing strategies, such as setting the unit price just below the previous lag price (e.g., $9.99 instead of $10), to create a perception of value and encourage purchasing.

# Market Timing: Tracking how quickly unit prices adjust to changes in lag prices can reveal how businesses respond to market shifts. Rapid adjustments might indicate a market where businesses are agile in responding to changes, while slower adjustments could signal more cautious pricing strategies.

# In summary, the relationship between unit price and lag price can offer valuable insights into pricing strategies, consumer preferences, market dynamics, and the overall health of a market or industry. It can help businesses make informed decisions about pricing, positioning, and market entry, and it can guide analysts in understanding the behavior of consumers and competitors within a specific context.





# if unit price and lag price is showing proportional relation or high positive correlation that means:

# #If unit price and lag price show a proportional relationship or a high positive correlation, it generally indicates that changes in the lag price tend to result in corresponding changes in the unit price. In other words, when the lag price increases (or decreases), the unit price also tends to increase (or decrease) by a similar or proportional amount. This relationship could imply several potential scenarios:

# Stable Pricing Environment: A high positive correlation between unit price and lag price may suggest a relatively stable pricing environment. Businesses might adjust their unit prices in direct response to changes in historical prices, aiming to maintain consistent profit margins or align with market trends.

# Price-Passing Behavior: This type of correlation could indicate that businesses tend to "pass on" changes in input costs, inflation, or other external factors directly to consumers. When the lag price rises, businesses increase the unit price to preserve their profit margins, and vice versa.

# Lack of Pricing Flexibility: High positive correlation might also suggest that businesses have limited pricing flexibility. They might feel constrained to adjust their unit prices closely in line with changes in lag prices, potentially due to factors like competitive pressures or consumer expectations.

# Consumer Expectations: If consumers have come to expect that unit prices will change in direct response to lag price changes, a high positive correlation could be a reflection of these expectations. This might lead to predictable consumer behavior in response to price changes.

# Market Competition: In a competitive market, a proportional relationship between unit price and lag price could indicate that businesses are closely monitoring each other's pricing strategies and quickly adjusting their prices to match competitors' moves.

# Cost-Pass Through: In industries where changes in input costs (e.g., raw materials, labor) strongly affect the final product price, a high positive correlation might indicate effective cost-pass through from input costs to consumers.

# Limited Pricing Power: A strong positive correlation might suggest that businesses lack the pricing power to deviate significantly from changes in historical prices. This could be due to intense competition, substitutability of products, or other market factors.

It's important to note that while a proportional relationship or high positive correlation between unit price and lag price can provide valuable insights, it's just one piece of the puzzle. Other factors, such as consumer behavior, market conditions, competition, and external influences, can also impact pricing decisions. Therefore, any interpretation should be considered within the broader context of the specific market or industry being analyzed.

```python
columns_to_remove_outliers = ['total_price', 'unit_price','avg_comp_diff','lag_price']
Q1 = df[columns_to_remove_outliers].quantile(0.25)
Q3 = df[columns_to_remove_outliers].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outlier_mask = ((df[columns_to_remove_outliers] >= lower_bound) & (df[columns_to_remove_outliers] <= upper_bound)).all(axis=1)
df_no_outliers = df[outlier_mask]
```

# VAR MODEL

```python
from statsmodels.tsa.api import VAR
from statsmodels.tsa.api import VAR
from statsmodels.tsa.stattools import adfuller
```

```python
def test_stationarity(timeseries):
    result=adfuller(timeseries,autolag='AIC')
    p_value=result[1]
    return p_value<=0.05
```

```python
print(test_stationarity(df['total_price']))
```

```python
#Train-test split
train_size = int(0.8 * len(df2))
train, test = df2[:train_size], df2[train_size:]

# Initialize and fit the VAR model
model = VAR(train)
model_fitted = model.fit()

# Make predictions
lag_order = model_fitted.k_ar
forecast = model_fitted.forecast(train.values[-lag_order:], steps=len(test))

# Convert forecasted values to a DataFrame
forecast_df = pd.DataFrame(forecast, index=test.index, columns=test.columns)

# Calculate RMSE
from sklearn.metrics import r2_score
r2_score = r2_score(test, forecast_df)

# Print RMSE score
print("r2_score:", r2_score)
```
